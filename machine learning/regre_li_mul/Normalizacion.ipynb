```python
def csvAdapted(dataset, modifyColumns, dropColumns):
    data = pd.read_csv(dataset, encoding='latin-1')
    columns = data.columns
    
    for index in modifyColumns:
        name = columns[index]
        value = list( data[name].unique() )
        data[name] = data[name].map( dict( zip( value,  [i for i in range( len(value) ) ] ) ) )

    data = data.drop([columns[index] for index in dropColumns ], axis=1)

    print(data.head())

    return data.to_numpy()

data = csvAdapted("datasets/ford.csv", [i for i in range(0, 8) if not i in [1, 2, 4, 6, 7, 8] ], [])
```

```python
data = pd.read_csv("dataset-tortuga.csv", encoding='latin-1')

UniqueData = [
    [ list( data.PROFILE.unique() ), "PROFILE" ],
]

DictData = []

for value, name in UniqueData:
    DictData.append(  [ dict( zip( value,  [i for i in range( len(value) ) ] ) ), name ])

newDataSet = pd.DataFrame()

newDataSet["HOURS_DATASCIENCE"] = data['HOURS_DATASCIENCE']
newDataSet["HOURS_BACKEND"] = data['HOURS_BACKEND']
newDataSet["HOURS_FRONTEND"] = data['HOURS_FRONTEND'] 
newDataSet["NUM_COURSES_BEGINNER_DATASCIENCE"] = data['NUM_COURSES_BEGINNER_DATASCIENCE'] 
newDataSet["NUM_COURSES_BEGINNER_BACKEND"] = data['NUM_COURSES_BEGINNER_BACKEND'] 
newDataSet["NUM_COURSES_BEGINNER_FRONTEND"] = data['NUM_COURSES_BEGINNER_FRONTEND'] 
newDataSet["NUM_COURSES_ADVANCED_DATASCIENCE"] = data['NUM_COURSES_ADVANCED_DATASCIENCE'] 
newDataSet["NUM_COURSES_ADVANCED_BACKEND"] = data['NUM_COURSES_ADVANCED_BACKEND'] 
newDataSet["NUM_COURSES_ADVANCED_FRONTEND"] = data['NUM_COURSES_ADVANCED_FRONTEND'] 
newDataSet["AVG_SCORE_DATASCIENCE"] = data['AVG_SCORE_DATASCIENCE'] 
newDataSet["AVG_SCORE_BACKEND"] = data['AVG_SCORE_BACKEND'] 
newDataSet["AVG_SCORE_FRONTEND"] = data['AVG_SCORE_FRONTEND'] 

for value, name in DictData:
    newDataSet[name] = data[name].map(value)

newDataSet = newDataSet.dropna()
newDataSet.to_csv("dataset-tortuga-2.csv", index=False)

newDataSet
```
